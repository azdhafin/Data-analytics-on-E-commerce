```python
# Transalating the product category names to English
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='pt', target='en')
cache = {}

def translate_category(name):
    if pd.isna(name):
        return name
    if name not in cache:
        readable = str(name).replace('_', ' ')
        cache[name] = translator.translate(readable)
    return cache[name]

df_products['product_category_name_english'] = df_products['product_category_name'].apply(translate_category)
```

```python
df_products['product_category_name_english'].value_counts().head(20)
```

```python
# Dropped the original category name column
df_products = df_products.drop(columns=['product_category_name'])
df_products.head(10)
```

```python
# checking for null values
df_products.isnull().sum()
```

```python
# dropping null values
df_products = df_products.dropna()
```

```python
# Saved changes to the file
df_products.to_csv('olist_products_dataset.csv', index=False)
```
